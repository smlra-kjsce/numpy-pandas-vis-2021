```python
#creating a pandas series from a list
my_list = [10,20,30]
series = pd.Series(my_list)

print(series)
print(series.index)
print(series.values)
```

```python
#creating a series from numPy Array
import numpy as np 
import pandas as pd

index=['a','b','c']
arr = np.array([10,20,30])

pd.Series(data=arr,index=index)
```

```python
#creating a series from dictionary
import pandas as pd 

d = {'a':10,'b':20,'c':30}
pd.Series(d)
```

```python
#Custom index
import pandas as pd 
ser1 = pd.Series([1,2,3,4], index=['USA','Germany','USSR','Japan'])
ser2 = pd.Series([1,2,3,4], index=['USA','Germany','Italy','Japan'])

print(ser1)
print(ser2)
#get the value of 'USA'
print(ser1['USA'])
```

```python
print(ser1+ser2)
```

```python
#Creating DataFrames Manually
import pandas as pd 
import numpy as np

df = pd.DataFrame([[1,2,3],
                    [3,4,5],
                    [5,6,7],
                    [7,8,9]])

df
```

```python
df = pd.DataFrame([[1,2,3],[3,4,5],[5,6,7],[7,8,9]])

print("Shape:",df.shape)
print("Index:", df.index)
df
```

```python
df2 = pd.DataFrame([[1,2,3],[3,4,5],[5,6,7],[7,8,9]],
                    index = ['a','b','c','d'], columns=['x','y','z'])

print("Shape:",df2.shape)
print("Index:",df2.index)

df2
```

```python
df.join(df2,how='outer')
```
